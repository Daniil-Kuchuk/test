```python
import math
n=len(s)

A = (((s-s.mean())**3).sum()/n)/s.std()**3
sigmaA = math.sqrt((6*(n-1))/((n+1)*(n+3)))

print(f'A = {A}')
print(f'|A| / sigmaA = {abs(A)/sigmaA}')
```

A > 0, следовательно скос вправо. |A| / sigmaA > 3, следовательно есть существенная асимметрия.

```python
n=len(s)

E = (((s-s.mean())**4).sum()/n)/s.std()**4-3
sigmaE = math.sqrt((24*n*(n-2)*(n-3))/((n-1)**2*(n+3)*(n+5)))

print(f'E = {E}')
print(f'|E| / sigmaE = {abs(E)/sigmaE}')
```

E > 0, следовательно распределение островершинно. |E| / sigmaE больше 3, значит есть отколнения. Е не существенно и не имеет смысла

```python
d1 = s
d2 = data['P5_2016']
g1 = np.histogram(data['P5_2016'])
display(g1)
st_group = pd.DataFrame()
data_group = []
new_Inter = []

for i in range(len(g1[0])):
    group = []
    temp = 0
    left_inter = g1[1][i]
    right_inter = g1[1][i+1]
    new_Inter.append(str(left_inter) + '-' + str(right_inter))
    for i in d1:
        if left_inter < i <= right_inter:
            group.append(d2[temp])
            temp += 1
    data_group.append(np.array(group))

st_group['Интервалы'] = pd.Series(new_Inter)
st_group['Частота'] = pd.Series([len(x) for x in data_group])
st_group['Внутрення средняя'] = pd.Series([(x.mean() if len(x) > 0 else 0) for x in data_group])
st_group['Внутренняя дисперсия'] = pd.Series([(x.std()**2 if len(x) > 0 else 0) for x in data_group])
display(st_group)
```

```python
from scipy.stats import t
def matrix_minor(arr, i, j):
    return np.delete(np.delete(arr,i,axis=0), j, axis=1)
def r_partial(arr, i,j):
    aij=(-1)*(i+j)*np.linalg.det(matrix_minor(np.array(arr), i,j))
    aii=np.linalg.det(matrix_minor(np.array(arr), i,i))
    ajj=np.linalg.det(matrix_minor(np.array(arr), j,j))
    return -aij/math.sqrt(aii*ajj)
def corr_partial(arr):
    res=np.copy(arr)
    for i in range(np.shape(arr)[0]):
        for j in range(np.shape(arr)[0]):
            if i!=j:
                res[i,j]=r_partial(arr, i,j)
            else:
                res[i,j]=-r_partial(arr, i,j)
    res=pd.DataFrame(res)
    res.index=arr.index;res.columns=arr.columns
    return res
MyData=pd.DataFrame({'1':d1,'2':d2}).corr()
display(corr_partial(MyData).iloc[1,0])
```

Найдем разность значений корреляционных матриц для нормированных и не нормированных значений. Если разность существенная по абсолютному значению, значит для данных двух параметров существует иная корреляционная зависимость. Если разница не существенная, поставим значение 0.

Как видно из матрицы,у нас нет существенных отклонений.

```python
func = (lambda x : (0 if abs(x)<10e-3 else x))
CorrR = corr-corr2
for col in CorrR.columns:
    s=CorrR[col]
    s=s.apply(func=func)
    CorrR[col]=s
CorrR.style.background_gradient(cmap='plasma')
```

```python
from sklearn.cross_decomposition import CCA
W=pd.DataFrame()
W["P1_2014"]=data["P1_2014"]
W["P1_2015"]=data["P1_2015"]
W["P1_2016"]=data["P1_2016"]
display(W)
P=data.iloc[:,2:4]
display(P)
scaler = preprocessing.StandardScaler().fit(W)
Ws=scaler.transform(W)
display(pd.DataFrame(Ws))
scaler = preprocessing.StandardScaler().fit(P)
Ps=scaler.transform(P)
display(pd.DataFrame(Ps))
```

```python
ca = CCA(n_components=3, scale=False, copy=True)
ca.fit(Ps,Ws)
U, V=ca.transform(Ps,Ws)
print(U)
print("--------------------------------------------")
print(V)
print(ca.x_weights_)
print(ca.y_weights_)
print(U[:,0])
print("-----------------------------------------")
print(np.dot(np.matrix(ca.x_weights_[:,0]), np.transpose(Ps)))
print(V[:,0])
print("-----------------------------------------")
print(np.dot(np.matrix(ca.y_weights_[:,0]), np.transpose(Ws)))
```

```python
display(np.corrcoef(U[:, 0], V[:, 0]))
display(np.corrcoef(U[:, 1], V[:, 1]))
display(np.corrcoef(U[:, 1], V[:, 0]))
display(np.corrcoef(V[:, 0], V[:, 1]))
```
