```python
#1
import random

lst = [random.randint(1, 10) for i in range(10)]
print(lst)
val = int(input('Введите число: '))


def f(arr, val):
    new_lst = []
    for item in arr:
        if item % val == 0:
            new_lst.append(item)
    return new_lst

print(f(lst, val))
```

```python
#2
import random
a=[random.randint(1, 10) for i in range(10)]
print(a)
def f(a):
    b=[]
    for i in range(len(a)):
        if (i+1) %5 == 0:
            b.append(a[i]*2)
        if (i+1) %3 != 0:
            b.append(a[i])        
    return b
print(f(a))
```

```python
#3
import random

a = [random.randint(1, 10) for i in range(10)]

print(a)

n = int(input('Введите число: '))


def f(a, n):
    sum = 0
    for i in a:
        if i == n:
            sum += 1
    return sum

print(f(a, n))
```

```python
#4
import random

a = [random.randint(1, 10) for i in range(12)]
print(a)

def f(arr):
    new_lst = []
    i = 0
    for _ in range(int( len(arr) / 3) ):
        a = arr[i:i+3]
        new_lst.extend(a[::-1])
        i += len(a)
    return new_lst

print(f(a))
```
